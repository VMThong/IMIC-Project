```python
#https://sharkcoder.com/data-visualization/seaborn-boxplot

mau_nhat='#eaeaf2'
font_color = '#525252'
font_name='calibri'
sns.set(style='whitegrid')
fig,ax=plt.subplots(figsize=(15,10),facecolor=mau_nhat)
ax=sns.boxplot(data=taxis,y='fare',x='passengers')

#Thay đổi font cho xtick và ytick hàng loạt, lý do khi cộng lại các xtick và ytick thì chứa trong 1 list,
# mà vì bản chất giống nhau nên khi xài set ta có thể thay đổi hàng loạt
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set(fontsize=16, color=font_color, fontname=font_name)

#Tạo title
title='MỨC THUẾ ÁP THEO SỐ LƯỢNG HÀNH KHÁCH'
fig.suptitle(title, y=.95, fontsize=22, color=font_color, fontname=font_name)

sub_title='Cho xe Taxi'
plt.title(sub_title, fontsize=18, pad=10, color=font_color, fontname=font_name)

sub_title1='Cho xe Taxi_1'
ax.set_title(sub_title1, fontsize=18, pad=10, color='r', fontname=font_name)
plt.subplots_adjust(top=0.87)

palette=[]
#Làm màu cho điểm ngoại lai tương ứng với các box
for i, box in enumerate(ax.artists):
    #Lấy màu box từ patch pat
    col = box.get_facecolor()
    plt.setp(ax.lines[i*6+5], mfc=col, mec=col)
    palette.append(col)

bien_lines = ax.get_lines()

categories = ax.get_xticks()

# thêm các nhãn vào median
for cat in categories:
    y = round(bien_lines[4+cat*6].get_ydata()[0],1) 
    ax.text(cat, y, f'{y}', ha='center', va='center', fontweight='semibold', size=12, color='white',
        bbox=dict(facecolor='#828282', edgecolor='#828282'))

# Thêm nhãn vào tứ phân vị
print(bien_lines[0].get_ydata())
print(bien_lines[1].get_ydata())
print(bien_lines[2].get_ydata())
print(bien_lines[3].get_ydata())
print(bien_lines[4].get_ydata())
print(bien_lines[5].get_ydata())

#Tạo một line nối các giá trị trung bình trong các boxplot lại
marker_tup=['o','o','o','o','o','o','o']

ax2=sns.lineplot(data=gb_paseen,y='fare',x='passengers',color='black',lw=2,markers=marker_tup,ms=10,style='passengers')
sns.lineplot(data=gb_paseen,y='fare',x='passengers',color='black',lw=2)

for z,tick in enumerate(ax.get_xticks()):
    ax.text(tick, mean[z]-1, f'{mean[z]}', ha='center', va='center', fontweight='semibold', size=12, color='white',
        bbox=dict(facecolor='#828282', edgecolor='#828282'))
    
ax.set_ylabel('Mức thuế',fontsize=20,fontname=font_name,fontweight='bold',ha='center',va='bottom',color='black')
ax.set_xlabel('Số lượng hành khách',fontsize=20,fontname=font_name,fontweight='bold',ha='center',va='top',color='black')
ax.set_ylim(0,40)


#Tạo legend chứa nhãn mean và số lượng hành khách
handles, labels = ax2.get_legend_handles_labels()

for z0,z in enumerate(handles):
    z.set(marker='o',linestyle='',markerfacecolor=palette[z0],markeredgecolor=palette[z0])

patch = matplotlib.patches.Patch(color='black', label='Manual Label',linestyle='-')

line=matplotlib.lines.Line2D([0], [0], label='Trung bình', color='k',marker='o')

handles.extend([line,patch])
ax2.legend(handles=handles[:], title="Số hành khách", bbox_to_anchor=(1.02, 1.0), loc='upper left')

handles
```

# Facet Grid

```python
import matplotlib.pyplot as plt
import numpy as np

def add_mean_line(data, var=None, **kws):
    
    # If no variable provided skip adding mean line
    if not var: return
    
    #Calculate mean for each group
    m = np.mean(data[var])
    #Get current axis
    ax = plt.gca()
    
    #add line at group mean
    ax.axvline(m, color='maroon', lw=3, ls='--')
    
    #annotate group mean
    x_pos=0.65
    if m > 5000: x_pos=0.2
    ax.text(x_pos, 0.7, f'mean={m:.0f}', 
            transform=ax.transAxes,   #transforms positions to range from (0,0) to (1,1)
            color='maroon', fontweight='bold', fontsize=12)
```

```python
g=sns.FacetGrid(data=titanic,col='alive',row='alone',height=6,hue='embarked',palette='prism',  margin_titles=True)
g.map_dataframe(add_mean_line,var='age',data=titanic)
g.map_dataframe(sns.histplot,x='age')
g.set_axis_labels('Độ tuổi','Tiền vé')
g.add_legend(title='hello')
g.set_xlabels(fontsize=20,fontname='arial',color='r',rotation=10);
```
